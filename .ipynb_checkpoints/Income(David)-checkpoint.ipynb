```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
overall = pd.read_csv("income.csv")
```

```python
overall = overall.sort_values("Estimate; Total:", ascending = False)
overall.head()
```

```python
neighByTotal = overall[['Neighborhood', 'Estimate; Total:']]
neighByTotal.head()
```

```python
neighByTotal.plot(x = "Neighborhood",y="Estimate; Total:", legend = None, kind = 'barh', figsize = (10,25))
plt.title('Neighborhoods by Income')
plt.xlabel('Income Over the Past 12 Months')
plt.ylabel("Neighborhood")
```

**Conclusion**

This is the graph and code work for the income metric. We choose this metric because income, at least in our heads, will likely correspond to better houses, lower crime, more parks, better maintained sidewalks, and more. 

Dataset: https://data.wprdc.org/dataset/pittsburgh-american-community-survey-2015-miscellaneous-data/resource/d159c6b3-f70a-439b-bd2b-0cb8735d76d4

As demonstrated by the graph, Shady Side (SS), Squirrel Hill South (SHS), and Brookline stand out from the rest. While SS and SHS are close in income, they both surpass every other neighborhood by at least 1000, and Brookline surpasses the others excluding the former two by over 1000 as well. This shows that SS and SHS are the best neighborhoods by this metric. 

It is worth noting the flaws in this dataset and thus graph. There was no provided information on how much each income is divided by. We can safely assume the average person in SS is not earning 7k per year, so it is likely that 7k is multiplied by 10 to make 70k per year. However, the exact number to us does not really matter since this is a comparative ranking, where we know at least that SS on average has higher income than SHS. We can thus safely say that SS, SHS, and Brookline win for this metric's best neighborhood.
